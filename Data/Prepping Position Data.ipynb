```python
positive_data = pd.read_csv('raw_data/gripper_pose (rev 1 - closest center).tsv', sep='\t')
negative_data = pd.read_csv('raw_data/negative_grasp_data.csv')
```

```python
res_pos_data = positive_data[(abs(positive_data.x)<200) & (abs(positive_data.y)<200) & (positive_data.z<500)]
```

```python
total_drop = 600
per_axis_drop = total_drop//3
top_x = res_pos_data.x.nlargest(n=per_axis_drop//2).iloc[-1]
bot_x = res_pos_data.x.nsmallest(n=per_axis_drop//2).iloc[-1]

top_y = res_pos_data.y.nlargest(n=per_axis_drop//2).iloc[-1]
bot_y = res_pos_data.y.nsmallest(n=per_axis_drop//2).iloc[-1]

top_z = res_pos_data.z.nlargest(n=per_axis_drop//2).iloc[-1]
bot_z = res_pos_data.z.nsmallest(n=per_axis_drop//2).iloc[-1]
```

```python
x_bool_series = (res_pos_data.x>bot_x) & (res_pos_data.x<top_x)
y_bool_series = (res_pos_data.y>bot_y) & (res_pos_data.y<top_y)
z_bool_series = (res_pos_data.z>bot_z) & (res_pos_data.z<top_z)
```

```python
res_pos_data = res_pos_data[x_bool_series & y_bool_series & z_bool_series]
```

```python
ids = list(set(negative_data.id.values).intersection(set(res_pos_data.id.values)))
```

```python
keep_pos_data = res_pos_data[res_pos_data.id.isin(ids)][['id','x','y','z','rx','ry','rz']]
keep_neg_data = negative_data[negative_data.id.isin(ids)]
```

```python
keep_pos_data.loc[:, 'x'] /= 1000
keep_pos_data.loc[:, 'y'] /= 1000
keep_pos_data.loc[:, 'z'] /= 1000
```

```python
def rotate_x(theta):
    R = np.array([[1, 0, 0],
                  [0,np.cos(theta),-np.sin(theta)],
                  [0,np.sin(theta), np.cos(theta)]])
    return R

def rotate_y(theta):
    R = np.array([[np.cos(theta),0,np.sin(theta)],
                  [0, 1, 0],
                  [-np.sin(theta),0, np.cos(theta)]])
    return R

def rotate_z(theta):
    R = np.array([[np.cos(theta), -np.sin(theta), 0],
                  [np.sin(theta),  np.cos(theta), 0],
                  [            0,              0, 1]])
    return R
```

```python
R1 = rotate_x(np.deg2rad(-102))
R2 = rotate_y(np.deg2rad(-0.9))
R3 = rotate_z(np.deg2rad(-135))
R = R3 @ R2 @ R1
p = np.array([[-35.3906], [28.6732], [22.75]])*1e-2
```

```python
frame = keep_pos_data
rot_mag = np.linalg.norm(frame.values[:,4:],axis=-1)%(2*np.pi)
rot = frame.values[:,4:]/rot_mag[...,np.newaxis]
pos = frame.values[:,1:4]
ids = frame.values[:,0].astype(np.int32)

pos = (R.T @ (pos.T - p)).T
rot = (R.T @ rot.T).T

final_pos_data = pd.DataFrame({'id': ids, 'x': pos[:,0], 'y': pos[:,1], 'z': pos[:,2], 'rx': rot[:,0], 'ry': rot[:,1], 'rz': rot[:,2], 'rot_mag':rot_mag})
```

```python
final_pos_data.to_csv('matched_pos_data.csv', index=False)
```

```python
frame = keep_neg_data
rot_mag = np.linalg.norm(frame.values[:,4:],axis=-1)%(2*np.pi)
rot = frame.values[:,4:]/rot_mag[...,np.newaxis]
pos = frame.values[:,1:4]
ids = frame.values[:,0].astype(np.int32)

pos = (R.T @ (pos.T - p)).T
rot = (R.T @ rot.T).T

final_neg_data = pd.DataFrame({'id': ids, 'x': pos[:,0], 'y': pos[:,1], 'z': pos[:,2], 'rx': rot[:,0], 'ry': rot[:,1], 'rz': rot[:,2], 'rot_mag':rot_mag})
```

```python
final_neg_data.sort_values('id').to_csv('matched_neg_data.csv', index=False)
```

```python
dx = (final_pos_data.x.max() - final_pos_data.x.min())/8
dy = (final_pos_data.y.max() - final_pos_data.y.min())/8
dz = (final_pos_data.z.max() - final_pos_data.z.min())/8
```

```python
def set_limits(iteration, base, step):
    if iteration==0:
        return -np.inf, base+step
    elif iteration==7:
        return base+7*step, np.inf
    else:
        lower = base + iteration*step
        higher= base + (iteration+1)*step
        return lower, higher

def voxel_choice(point):
    x,y,z = point
    for i in range(8):
        lower_x, higher_x = set_limits(i, final_pos_data.x.min(), dx)
        in_x = lower_x < x < higher_x
        if not in_x:
            continue
        for j in range(8):
            lower_y, higher_y = set_limits(j, final_pos_data.y.min(), dy)
            in_y = lower_y < y < higher_y
            if not in_y:
                continue
            for k in range(8):
                lower_z, higher_z = set_limits(k, final_pos_data.z.min(), dz)
                in_z = lower_z < z < higher_z
                if not in_z:
                    continue
                return k,j,i
```

```python
ids    = final_pos_data.id.values
points = final_pos_data[['x','y','z']].values
```

```python
coords = np.zeros((points.shape[0],3), dtype=np.int32)
for i,point in enumerate(points):
    coord = voxel_choice(point)
    coords[i] = coord
```

```python
coords = np.zeros((points.shape[0],3), dtype=np.int32)
for i,point in enumerate(points):
    coord = voxel_choice(point)
    coords[i] = coord
```

```python
np.savetxt('negative_position_cube.csv', np.concatenate([ids[...,np.newaxis],coords], axis=-1), fmt='%d', delimiter=',', header='id,z,y,x')
```
